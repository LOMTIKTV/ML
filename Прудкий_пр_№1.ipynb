```python
#1.	Створіть одновимірний масив (вектор) з числами, які є кубами натуральних чисел, починаючи з 1, та виведіть його значення.
import numpy as np

a = np.arange(1, 11) ** 3

# Виведення значень вектора
print(a)
```

```python
#2.	Створіть двовимірний масив (матрицю) розміром 3x3, 
#заповніть його значеннями факторіалів кожного числа плюс його попереднього числа, починаючи з 1, та виведіть його значення.

import math
matrix = np.zeros((3, 3), dtype=int)
for i in range(3):
    for j in range(3):
        current_number = i * 3 + j + 1
        factorial = np.math.factorial(current_number)
        previous_number = current_number - 1 if current_number > 1 else 0
        matrix[i, j] = factorial + previous_number

print(matrix)
```

```python
#3.Створіть масив розміром 9x9, заповніть його випадковими цілими числами від -5 до 5 та знайдіть суму модулів його елементів.

matrix = np.random.randint(-5, 6, size=(9, 9))
sum_of_abs_elements = np.sum(np.abs(matrix))
print("Масив:")
print(matrix)
print("Сума модулів елементів масиву:", sum_of_abs_elements)
```

```python
#4.Створіть масив розміром 3x3, заповніть його випадковими дійсними числами 
#в діапазоні від 0 до 1 та знайдіть середнє геометричне всіх елементів.

matrix = np.random.rand(3, 3)
geometric_mean = np.prod(matrix) ** (1 / matrix.size)
print("Масив:")
print(matrix)
print("Середнє геометричне всіх елементів масиву:", geometric_mean)
```

```python
#5.Створіть два одновимірних масиви розміром 5, заповніть їх випадковими дійсними 
#числами від 0 до 1 та виконайте на них поелементні операції додавання, віднімання та множення.
a = np.random.rand(5)
b = np.random.rand(5)

print("a = ", a)
print("b = ", b)

print("a + b = ", a + b)
print("a - b = ", a - b)
print("a * b = ", a * b)
```

```python
#6.Створіть два вектори розміром 6, заповніть їх довільними дійсними числами та знайдіть їх скалярний добуток.
a = np.random.rand(6)
b = np.random.rand(6)

dot_a_b = np.dot(a, b)

print("a = ", a)
print("b = ", b)
print("Скалярний добуток векторів дорівнює:", dot_a_b)
```

```python
#8.	Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю.
a = np.random.randint(1, 16, size=(3, 3))
print("Вихідна матриця:")
print(a)

# перевірка, чи можна знайти обернену матрицю
if np.linalg.det(a) == 0:
    print("Матриця не має оберненої матриці.")
else:
    # знаходження оберненої матриці
    inv_a = np.linalg.inv(a)
    print("Обернена матриця:")
    print(inv_a)
```

```python
#10.Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими дійсними числами від 0 до 1,
#а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму елементів отриманої матриці.
matrix = np.random.rand(3, 4)
vector = np.random.rand(4)
result = np.dot(matrix, vector)
sum_result = np.sum(result)

print("Матриця:")
print(matrix)
print("Вектор:")
print(vector)
print("Результат перемноження:")
print(result)
print("Сума елементів отриманої матриці:", sum_result)
```

```python
#12.Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 
matrix1 = np.random.randint(1, 11, (3, 3))
matrix2 = np.random.randint(1, 11, (3, 3))
diagonal_sum = np.trace(matrix2)
print("Перша матриця:")
print(matrix1)
print("Друга матриця:")
print(matrix2)
print("Сума елементів діагоналі другої матриці:", diagonal_sum)
```

```python
#13.Створіть дві матриці розміром 2x2, заповніть їх випадковими цілими числами в діапазоні від -10 до 10
#та знайдіть добуток їх елементів.
matrix1 = np.random.randint(-10, 11, (2, 2))
matrix2 = np.random.randint(-10, 11, (2, 2))
result = np.prod(matrix1) * np.prod(matrix2)

# Виведення матриць та добутку їх елементів
print("Перша матриця:")
print(matrix1)
print("Друга матриця:")
print(matrix2)
print("Добуток елементів матриць:", result)
```

```python
#15.Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та 
#знайдіть вектор-стовпчик,що містить кількість від'ємних елементів в кожному рядку матриці.

matrix = np.random.uniform(-1, 1, (4, 4))
vid = np.sum(matrix < 0, axis=1)
vid1 = vid.reshape(-1, 1)
print("Матриця:")
print(matrix)
print("Вектор-стовпчик з кількістю від'ємних елементів в кожному рядку:")
print(vid1)
```
